```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 50)
```

<a name='1'></a>
### Colores

```python
# Diferentes maneras de asignar un color
plt.plot(x, np.cos(x - 0), color='blue')        # Por nombre
plt.plot(x, np.cos(x - 1), color='g')           # Código corto(rgbcmyk)
plt.plot(x, np.cos(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.cos(x - 3), color='#FFDD44')     # Hex code (RRGGBB de 00 a FF)
plt.plot(x, np.cos(x - 4), color=(1.0,0.2,0.3)) # tupla RGB, valores 0 to 1
plt.plot(x, np.cos(x - 5), color='chartreuse')  # colores HTML
```

```python
# Cada estilo tiene una lista de colores predefinidos
plt.style.use('seaborn-white') # seaborn-white, ggplot
plt.plot(x, np.cos(x - 0), color='C0')        
plt.plot(x, np.cos(x - 1), color='C1')           
plt.plot(x, np.cos(x - 2), color='C2')        
plt.plot(x, np.cos(x - 3), color='C3')     
plt.plot(x, np.cos(x - 4), color='C4') 
plt.plot(x, np.cos(x - 5), color='C5')
```

<a name='2'></a>
### Tipos de líneas

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# Se pueden usar códigos
plt.plot(x, x + 4, linestyle='-')  
plt.plot(x, x + 5, linestyle='--') 
plt.plot(x, x + 6, linestyle='-.') 
plt.plot(x, x + 7, linestyle=':')

# Se puede usar 'ls' en vez de linestyle
plt.plot(x, x + 8, ls='-')  # solid
plt.plot(x, x + 9, ls='--') # dashed
plt.plot(x, x + 10, ls='-.') # dashdot
plt.plot(x, x + 11, ls=':');  # dotted

# Se puede usar una tupla para una definición más fina
plt.plot(x, x + 12, ls = (0, (1, 10)))
plt.plot(x, x + 13, ls = (0, (3, 10, 1, 10)))
plt.plot(x, x + 14, ls = (0, (3, 5, 1, 5, 1, 5)))
```

<a name='3'></a>
### Marcadores

```python
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p']:
    r = np.random.rand(2,10)
    plt.plot(r[0], r[1], marker, label=" '{}'".format(marker))
plt.legend()
plt.xlim(-0.15,1.5)
plt.ylim(-0.15,1.15)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Marcadores aleatorios')
```

```python
plt.plot(x, np.sin(x), 'p', color='C0',
         ls = '-', linewidth=2,
         markersize=12, 
         markerfacecolor='C2',
         markeredgecolor='gray',
         markeredgewidth=0.75,
         fillstyle='right')
plt.ylim(-1.2, 1.2);
```

<a name='4'></a>
### Texto en las gráficas

```python
x_d, y_d = 1, 5
x_a, y_a = 0.5, 0.1
x_f, y_f = 0.5, 0.5

fig = plt.figure(figsize=(6.0, 4.0))
plt.axis([0, 10, 0, 10])
ax = plt.gca()

plt.text(x_d, y_d, ". Data: ({x_d}, {y_d})", transform=ax.transData) # Default
plt.text(x_a, y_a, ". Axes: ({x_a}, {y_a})", transform=ax.transAxes)
plt.text(x_f, y_f, ". Figure: ({x_f}, {y_f})", transform=fig.transFigure)

plt.text(8, 8, "Direction", ha="center", va="center", rotation=45, size=15,
         bbox=dict(boxstyle="larrow,pad=0.3", fc="C6", ec="k", lw=1))
```

<a name='5'></a>
### Anotaciones sobre las gráficas

```python
x = np.linspace(0, 20, 1000)

fig = plt.figure(figsize=(6.0,4.0))
ax = plt.axes()
ax.axis('equal')

plt.plot(x, np.cos(x))

xmax = 2 * np.pi
ymax = np.cos(xmax)
xmin = 5 * np.pi
ymin = np.cos(xmin)

plt.annotate('Máximo local $\Delta$', xy=(xmax, ymax), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Mínimo local', xy=(xmin, ymin), xytext=(2, -6),
            arrowprops=dict(edgecolor='k', arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"))
```

### Leyendas

```python
plt.plot(x, np.cos(x - 0), color='blue', label='azul')
plt.plot(x, np.cos(x - 2), color='0.75', label='Gray scale = 0.75') 
plt.plot(x, np.cos(x - 4), color=(1.0,0.2,0.3), label='RGB =(1.0,0.2,0.3)') 
plt.plot(x, np.cos(x - 6), color='chartreuse', label='HTML = chartreuse') 

plt.legend()
```

```python
plt.plot(x, np.cos(x - 0), color='blue', label='azul') 
plt.plot(x, np.cos(x - 2), color='0.75', label='Gray scale = 0.75')
plt.plot(x, np.cos(x - 4), color=(1.0,0.2,0.3), label='RGB =(1.0,0.2,0.3)') 
plt.plot(x, np.cos(x - 6), color='chartreuse', label='HTML = chartreuse') 

plt.legend(loc='upper right')
```

```python
plt.plot(x, np.cos(x - 0), 'o', color='blue', label='azul') 
plt.plot(x, np.cos(x - 2), '-s', color='0.75', label='Gray scale = 0.75')
plt.plot(x, np.cos(x - 4), '--p', color=(1.0,0.2,0.3), label='RGB =(1.0,0.2,0.3)') 
plt.plot(x, np.cos(x - 6), '-*', color='chartreuse', label='HTML = chartreuse') 

plt.legend(loc='best', bbox_to_anchor=(0.95, 0.5, 0.5, 0.5))
```

```python
plt.plot(x, np.cos(x - 0), color='blue', label='$\cos(x)$') 
plt.plot(x, np.cos(x - 2), color='0.75', label='$\Delta x$') 
plt.plot(x, np.cos(x - 4), color=(1.0,0.2,0.3), label='rojo') 
plt.plot(x, np.cos(x - 6), color='chartreuse', label='verde')  

plt.legend(loc='best', bbox_to_anchor=(1.05, 0.5, 0.5, 0.5), ncol=2)
```
