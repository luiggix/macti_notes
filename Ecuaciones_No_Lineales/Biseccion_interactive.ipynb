```python
import os
import numpy as np
import matplotlib.pyplot as plt
import gutils as vis
%matplotlib inline

# funcion a graficar y de la cual se busca la raiz
def f(x):
  return (x**2)-1

x = np.arange(-0.25, 2, 0.1)
fx = [f(i) for i in x]
raiz,sucesion = vis.biseccion(f,0.001,1000,-0.5,2)

par = [{'title':'$f(x)$', 
        'xlabel':'$x$',
        'ylabel':'$y$'}]
graf = vis.planoCartesiano(par=par)
#funcion a grficar
graf.plot(x=x,y=fx)
#ejes
graf.plot(x = [-0.25, 2], y = [0, 0], par={'ls':'-', 'lw':1, 'color':'black'})
graf.plot(x = [0, 0], y = [-1, 2.5], par={'ls':'-', 'lw':1, 'color':'black'})

# iteracion 1
graf.annotate(par={'s':'$x_1$ = '+str(sucesion[0]), 'xy':(sucesion[0], 0), 'xycoords':'data',
                   'xytext':(0.5,-0.5), 'fontsize':12,
                   'textcoords':'data', 'ha':'center', 'va':'bottom',
                   'bbox':{'boxstyle':'round','pad':0.4, 'fc':'blue', 'alpha':0.25},
                   'arrowprops':{'arrowstyle':'->', 'connectionstyle':'arc3,rad=0.5', 'color':'black'}}) 
# iteracion 2
graf.annotate(par={'s':'$x_2$ = '+str(sucesion[1]), 'xy':(sucesion[1], 0), 'xycoords':'data',
                   'xytext':(1.5,-0.5), 'fontsize':12,
                   'textcoords':'data', 'ha':'center', 'va':'bottom',
                   'bbox':{'boxstyle':'round','pad':0.4, 'fc':'blue', 'alpha':0.25},
                   'arrowprops':{'arrowstyle':'->', 'connectionstyle':'arc3,rad=0.5', 'color':'black'}}) 
# iteracion n
graf.annotate(par={'s':'$x_n$ = '+str(sucesion[-3]), 'xy':(sucesion[-3], 0), 'xycoords':'data',
                   'xytext':(1.0,-1.0), 'fontsize':12,
                   'textcoords':'data', 'ha':'center', 'va':'bottom',
                   'bbox':{'boxstyle':'round','pad':0.4, 'fc':'blue', 'alpha':0.25},
                   'arrowprops':{'arrowstyle':'->', 'connectionstyle':'arc3,rad=0.5', 'color':'black'}}) 
# raiz
graf.scatter(x=[raiz],y=[0.],par={'c':'red','marker':'o','s':75, 'zorder':5})
graf.annotate(par={'s':'La raÃ­z de f es en x = '+str(raiz), 'xy':(raiz, 0), 'xytext':(-50,50), 'fontsize':15,
            'textcoords':'offset points', 'ha':'center', 'va':'bottom',
            'bbox':{'boxstyle':'round','pad':0.2, 'fc':'yellow', 'alpha':0.3},
            'arrowprops':{'arrowstyle':'->', 'connectionstyle':'arc3,rad=0.5', 
                           'color':'black'}})

graf.show()
```
