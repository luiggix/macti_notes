```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import interpoladores 
import macti.visual as mvis
```

```python
L = 4000.0  # Profundidad máxima
N = 99                      # Número de incógnitas
dx = L / (N+1)              # Distancia entre nodos
z = np.linspace(0, L, N+2)  # Coordenada z de la profundidad
```

```python
dat_ini = pd.read_csv('T_inicial.csv')
```

```python
plt.plot(dat_ini.Temp, dat_ini.Depth, '.--', lw = 0.75, label='Inicial')
plt.gca().invert_yaxis()
plt.xlabel('Temperatura [$^o$C]')
plt.ylabel('Profundidad [km]')
plt.legend()
plt.show()
```

```python
x, y, z_i = np.array(dat_ini.Depth[0:12]), np.array(dat_ini.Temp[0:12]), z[0:40]
```

```python
_, _, T_li = interpoladores.spline_lineal(x, y, z_i)
```

```python
plt.plot(y, x, '.--', lw = 0.75, label='Inicial')
plt.plot(T_li, z_i, '.', label = 'Interpolación lineal')
plt.gca().invert_yaxis()
plt.xlabel('Temperatura [$^o$C]')
plt.ylabel('Profundidad [km]')
plt.legend()
plt.show()
```

```python
d_q, s_q, T_ci = interpoladores.spline_quad(x, y, z_i)
```

```python
plt.plot(y, x, '.--', lw = 0.75, label='Inicial')
plt.plot(T_ci, z_i, '.', label = 'Interpolación lineal')
plt.plot(s_q, d_q)
plt.gca().invert_yaxis()
plt.xlabel('Temperatura [$^o$C]')
plt.ylabel('Profundidad [km]')
plt.legend()
plt.show()
```
