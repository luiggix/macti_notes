<a name='1-1'></a>
### Definicin y producto punto

Un producto interno o producto interior, sobre un espacio vectorial (para el caso de este notebook ser $R^n$), es una operacion que asigna a cada par de vectores u y v un escalar tal que <u|v> y cumple con las siguinetes propiedades:

* <v|v> $\geq 0$
* <v|v> = 0 si y solo si v =0
* <u|v+w> = <u|v> + <u|w>
* Sea $\alpha \in R$ entonces < $\alpha$ v|w> = $\alpha$<v|w>
* <v|w>= $\bar{<w|v>}$ (Esta ltima propiedad se usa con los nmeros complejos, por lo que para fines de este notebook no se usar pero es importante saber de su existencia)

En el caso de $ R^n$ el porducto interior es mejor conocido como producto punto y est definido como $\vec{a} \cdot \vec{b} = ||\vec{a}||||\vec{b}||cos $ donde $||\vec{a}||$ y $||\vec{b}||$ son la magnitud del vector $\vec{a}$ y $\vec{b}$ respectivamente y $$ el ngulo entre ambos.



Ten en cuenta que $\textbf{el producto punto de dos vectores es un escalar, no un vector.}$ Eso significa, por ejemplo, que no tiene sentido preguntar a qu es igual $\vec{a} \cdot \vec{b} \cdot \vec{c} $.

Una vez que hayamos evaluado $\vec{a} \cdot \vec{b}$
 y obtenido un nmero, terminaramos por tratar de tomar el producto punto entre un nmero y un vector.

 Otra forma equivalente y ms sencilla de calcular el producto punto es conociendo los coeficientes de sus componentes; de modo que el produto punto ser la suma de las multiplicaciones de sus componentes.

Supongamos que tenemos a los vectores 

$x = (x_1,x_2,...,x_n)$

$y = (y_1,y_2,...,y_n) $


Tenemos que el producto interior entonces ser:

$\sum_{i=1}^n x_iy_i$

Ahora para visualizar mejor el porducto punto vemos un ejemplo.

## Ejemplo de producto punto en $R^n$
Si tenemos los vectores 
$x = (3,2)$
$y = (5,6)$
 Entonces el producto punto ser $\alpha = (3)(5)+(2)(6)= 20$

Pero lo que realmente nos interesa es el valor lmite de esta suma conforme el tamao de esos pasos es ms y ms pequeo. Para ello haremos uso de la integral ya que toma en cuenta las contibuciones de las trayectorias infinitesimales y se ver de la forma:

$_C \vec{F}\cdot \vec{ds}$
Similar a una integral de lnea en un campo escalar pero ahora consideramos el pequeo paso $\vec{ds} $ como un vector, no como una longitud.


$\textbf{Nota}$: la cantidad que sumamos conforme caminamos a lo largo de C no es todo el valor de $\vec{F_g}$ en cada punto, sino la componente de $\vec{F_g}$ que apunta en la misma direccin que el vector $\vec{ds}$. Es decir, la componente de la fuerza en la direccin de la curva.

En el ejemplo anterior, el campo vectorial gravitacional es constante. En todos lados, la gravedad apunta directo hacia abajo con la misma magnitud. Para la mayora de las integrales de lnea en un campo vectorial, los vectores del campo varan en diferentes puntos del espacio, por lo que el valor de su producto punto con $d\vec{s}$ cambia.

Analicemos qu est sucediendo.La integral de lnea en s misma la escribimos como:

$\int_C \vec{F(r})\cdot d\vec{r} = \int_a^b \vec{F(r}(t))\vec{r'}(t) dt$

Donde:
$\vec{F}$ es uncampo vectorial, que asocia cada punto en el espacio con un vector. Lo puedes pensar como un campo de fuerzas.

C es una curva a travs del espacio.

$\vec{r}(t)$ es una funcin vectorial que parametriza la curva C, donde $a \leq t \leq b$.

$\vec{r'}(t)$ es la derivada de $\vec{r}$ que representa el vector velocidad de una partcula cuya posicin esta dada por $\vec{r}(t)$ mientras t se incrementa a razn constante. Cuando multiplicas este vector por un pequeo paso en el tiempo, dt, obtienes el vector de desplazamiento, que se puede  pensar como un pequeo paso a lo largo de la curva. Tcnicamente, es un pequeo paso en la direccin tangente a la curva.

$\textbf{Qu representa el producto punto $\vec{F(r(t))}\cdot\vec{r}'(t)dt$?}$
 
$\ \ $a) La razn con la cual la fuerza cambia respecto a t .
 
$\ \ $b) La componente de $\vec{F(r)}$ que paunta en la misma direccin que el peuqueo paso a lo largo de la curva en el punto $\vec{r}(t)$, nultiplicada por el tamao de ese paso.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Creamos una funcin que dibuja una curva y un vector en el punto especificado
def plot_vector(x0, y0):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x)
    dx = np.cos(x0)
    dy = np.sin(x0)
    plt.plot(x, y)
    plt.quiver(x0, y0, dx, dy, scale=5)
    plt.show()

# Creamos un widget deslizador para el ngulo del vector
angle_slider = widgets.FloatSlider(value=0, min=0, max=2*np.pi, step=0.1, description='Angle')

# Creamos una funcin de controlador de eventos que llama a la funcin de trazado de vectores con los valores del widget
def on_value_change(change):
    angle = change['new']
    x0 = angle
    y0 = np.sin(x0)
    plot_vector(x0, y0)

# Conectamos la funcin de controlador de eventos con el widget deslizador
angle_slider.observe(on_value_change, names='value')

# Mostramos el widget deslizador
display(angle_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# Creamos una funcin para el campo vectorial
def vector_field(x, y):
    dx = y
    dy = -x
    return dx, dy

# Creamos un widget para ajustar el tamao del vector
vector_scale = widgets.FloatSlider(
    value=1,
    min=0.1,
    max=3.0,
    step=0.1,
    description='Escala:',
    readout_format='.1f',
)

# Creamos una figura y ejes para graficar
fig, ax = plt.subplots(figsize=(5,5))

# Creamos una malla de puntos para graficar el campo vectorial
x, y = np.meshgrid(np.linspace(-3,3,10), np.linspace(-3,3,10))
dx, dy = vector_field(x, y)

# Graficamos el campo vectorial
ax.quiver(x, y, dx, dy, scale_units='xy', angles='xy', scale=1)

# Creamos una funcin que actualiza el tamao del vector
def update_vector_scale(change):
    ax.clear()
    ax.quiver(x, y, dx, dy, scale_units='xy', angles='xy', scale=change['new'])

# Conectamos el widget a la funcin de actualizacin
vector_scale.observe(update_vector_scale, names='value')

# Mostramos el widget y la figura
widgets.VBox([vector_scale, fig.canvas])
```
