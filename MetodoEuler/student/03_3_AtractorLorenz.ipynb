El atractor de Lorenz está definido por las siguientes ecuaciones:

$$
\dfrac{dx}{dt} = \sigma(y − x),\qquad
$$
$$
\dfrac{dy}{dt} = x(\rho − z) − y,\qquad
$$
$$
\dfrac{dz}{dt} = xy − \beta z
$$
para $\sigma = 10$, $\beta = 8/3$, $\rho = 28$.
