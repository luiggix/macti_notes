```python
import pandas as pd
import os
import shutil
```

```python
# Create the connection to the database
from nbgrader.api import Gradebook

# Creación del directorio submitted ----------------------
try:
    os.mkdir('submitted')
except FileExistsError:
    print('El directorio ya existe, continua sin problemas\n')
except:
    print('Ocurrió un error\n')
else:
    print('El directorio fue creado sin problemas\n')
#---------------------------------------------------------    

course_name = 'courseid_38_participants.csv'

with Gradebook('sqlite:///gradebook.db') as gb:
    
    student = pd.read_csv(course_name)
    
    for i, (n, s, e) in enumerate(zip(student['First name'], 
                       student['Surname'], 
                       student['Email address'])):
        student_id = e.split(sep='@')[0]
        student_name = n + ' ' + s
        
        # Creación del directorio de cada estudiante
        os.makedirs('submitted/' + student_id, exist_ok=True)
        print('({:02d}) id : {} \nNombre: {} \nemail: {} \n\n'.format(i+1, student_id, student_name, e))
        
        # Agregamos al estudiante a la base de datos
        gb.add_student(student_id, first_name=n, last_name=s, email=e)
```

```python
print(os.getcwd())

assigment_moodle_path = '../entregas/'

course_name = 'courseid_38_participants.csv'
student = pd.read_csv(course_name)

assigment = 'cond_1D'
notebook_list = ['conduccion1d_interactivo.ipynb']

submitted = {}

entregas = os.listdir(assigment_moodle_path)
for i in entregas:
    nombre = i.split(sep='_')[0]
    submitted[nombre] = ''
    
    # Obtiene el nombre completo del estudiante, lo busca en el dataframe 'student', 
    # una vez encontrado obtiene el email, luego lo convierte a una lista y finalmente
    # extrae la cadena del email del elemento 0
    email = list(student[student['First name'] + ' ' + student['Surname'] == nombre]['Email address'])[0]

    # Obtiene el student_id que tambien es el directorio donde se deben guardar las
    # entregas del estudiante.
    student_id = email.split(sep='@')[0]

    # Creamos el directorio de cada estudiante correspondiente al ejercicio
    student_nbgrader_path = 'submitted/' + student_id + '/' + assigment + '/'
    os.makedirs(student_nbgrader_path, exist_ok=True)
    
    print('dir {} \nNombre {} \n id {}'.format(i, nombre, student_id))

    student_moodle_path = assigment_moodle_path + i +'/'
    student_notebooks = os.listdir(student_moodle_path)
    
    os.rename(student_moodle_path + student_notebooks[0], 
              student_nbgrader_path + notebook_list[0])
```
