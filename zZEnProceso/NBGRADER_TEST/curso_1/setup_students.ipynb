```python
# Create the connection to the database
from nbgrader.api import Gradebook

# Creación del directorio submitted ----------------------
try:
    os.mkdir('submitted')
except FileExistsError:
    print('El directorio ya existe, continua sin problemas\n')
except:
    print('Ocurrió un error\n')
else:
    print('El directorio fue creado sin problemas\n')
#---------------------------------------------------------    

course_name = 'courseid_38_participants.csv'

with Gradebook('sqlite:///gradebook.db') as gb:
    
    student = pd.read_csv(course_name)
    
    for i, (n, s, e) in enumerate(zip(student['First name'], 
                       student['Surname'], 
                       student['Email address'])):
        id_est = e.split(sep='@')[0]

        # Creación del directorio de cada estudiante
        os.makedirs('submitted/' + id_est, exist_ok=True)
        print('({:02d}) id : {} \nNombre: {} \nemail: {} \n\n'.format(i+1, id_est, n + ' ' + s, e))
        
        # Agreamos al estudiante a la base de datos
        gb.add_student(id_est, first_name=n, last_name=s, email=e)
```

```python
import shutil
import os

"Move submissions to a dir called `submitted` directory. So, nbgrader can recognize them for grading."

# Path of directory containing students submissions initially (arbitrary dir)

arbit_dir = 'arbitrary/'### First Create a dir called 'submitted` in course_id dir

os.mkdir('submitted')

subm_dir = 'submitted/'

# Get files in arbitrary dir containing students submissions
files = os.listdir(arbit_dir)

# Move all files (submissions) from arbitrary directory to submitted dir
for f in files:
    shutil.move(arbit_dir+f, subm_dir)
```
